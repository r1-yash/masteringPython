Decorators

Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

```python
## function copy
def welcome():
    return "Welcome to the advanced python course"

welcome()
```

```python
#we can copy a funciton
wel=welcome
print(wel())
del welcome
print(wel())
```

```python
#closure functions are functions with sub functions 

def main_welcome(msg):
   
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
def main_welcome(func,lst):
       
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(func(lst))
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
### Decorator
def main_welcome(func):
   
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

so what does my decorator allows me to do is use a sub class by using @base_class_name 

```python
def coure_introduction():
    print("This is an advanced python course")

coure_introduction()
```

```python
main_welcome(coure_introduction)
#so here we pass the subclass as argument to main class
```

```python
@main_welcome
def coure_introduction():
    print("This is an advanced python course")

#but here we simply use @main_class and it automatically uses the main class
```

```python
## Decorator

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```

```python
## Decorators WWith arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```
