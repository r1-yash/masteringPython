Generators 

Generators are a simpler way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

```python
def square(n):
    for i in range(n):
        yield i**2 #here we would generally put return but what will it do, return us just 1 element
        
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python

a=square(3)
a
```

```python

def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
```

```python
next(gen)
```

```python

for val in gen:
    print(val)
```

#So the practical implementation ofthis is reading large files, and we dont need to save or load the file into memory and we can process one line at a time

```python
def read_large_file(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line
```

```python
file_path = 'large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
