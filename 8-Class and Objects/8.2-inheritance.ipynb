
Inheritance In Python

Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes and methods from another class. This lesson covers single inheritance and multiple inheritance, demonstrating how to create and use them in Python.



```python
class car():

    def __init__(self, windows, door, engine):
        self.windows=windows
        self.door=door
        self.engine=engine

    def rideType(self):
        print(f"The driver is riding a car with {self.engine} engine")
    



```

```python
bmw = car(4, 4, 'V6')
```

```python
bmw.rideType()
```

```python
class Tesla(car):
    
    def __init__(self, windows, door, engine, selfdriving):
        super().__init__(windows, door, engine)
        self.selfdriving=selfdriving
        #super allows me to inherit from one class 

    def doesSelfDrive(self):
        print(f"Can tesla selfdrive: {self.selfdriving} ")
```

```python
teslaX = Tesla(4,4,"electric", True)
teslaX.doesSelfDrive()
```

```python

### Multiple Inheritance
## When a class inherits from more than one base class.
## Base class 1
class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print("Subclass must implement this method")

## BAse class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner


##Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} say woof"
    

## Create an object
dog=Dog("Buddy","yash")
print(dog.speak())
print(f"Owner:{dog.owner}")
```
