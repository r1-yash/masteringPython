```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
even_or_odd(25)
```

```python

## Default Parameters

def greet(name="Default name I can give"):
    print(f"Hello {name} Welcome To the paradise")

greet("Yash")
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Yash")

```

```python

### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Yash",age="21",country="India")
```

```python

def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Yash",name="Yash",age="21",country="India")
#during runtime it automatically fugures out positional and keyword pair arguments 
```
