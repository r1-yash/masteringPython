```python
def pass_checker(password):

    if len(password) > 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char in '!@#$%' for char in password):
        return False
    return True
```

```python

pass_checker('chicken@3N')
```

```python
def calculate_total(cart):
    cost = 0
    for i in cart:
        cost += i['price'] * i['quantity']

    return cost
```

```python
cart=[
    {'name':'Apple','price':0.5,'quantity':4},
    {'name':'Banana','price':0.3,'quantity':6},
    {'name':'Orange','price':0.7,'quantity':3}

]

total_cost=calculate_total(cart)
print(total_cost)
```

```python
def ifPalindrome(s):
    sn = s
    stored_sn = ''.join(reversed(sn))
    return stored_sn==s
```

```python
ifPalindrome('acababbb')
```

```python
def altifPalindrome(s):
    s=s.lower().replace(" ", "")
    return s==s[::-1]
#[start : stop : step]
```

```python
altifPalindrome("abccbad")
```

```python
def factorial(n):
    if(n==0): return 1

    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
def wordCount(file_path):
    countWord = {}
    with open(file_path, 'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word=word.lower().strip('.,!?;:"\'')
                countWord[word] = countWord.get(word,0)+1
    return countWord


file_path = 'sample.txt'
word_freq = wordCount(file_path)
print(word_freq)

```
